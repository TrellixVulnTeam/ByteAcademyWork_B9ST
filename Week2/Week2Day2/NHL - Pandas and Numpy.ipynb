```python
def name_split(name):
    name = name.split('\\')[0]
    return name 

# name_split = lambda name: name.split("\\")[0]
```

```python
"""LAMBDA FUNCTION"""

lambda arguments: expression

square = lambda x: x * x

# same as 

def square(x):
    return x*x

```

```python
df['Player'] = df.apply(lambda row: name_split(row['Player']), axis=1)
```

```python
def name_strip(name):
    name = name.strip('*+\\')
    return name

#name_strip = lambda name: name.strip('*+')
```

```python
df['Player'] = df.apply(lambda row: name_strip(row['Player']), axis=1)
```
