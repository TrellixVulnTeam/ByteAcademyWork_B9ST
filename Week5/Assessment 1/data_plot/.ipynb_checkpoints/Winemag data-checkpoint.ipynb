```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("winemag-data-130k-v2.csv")

data.head()

```

```python
#* Load the Winemag csv data into a Pandas dataframe and create a scatter plot with x = price and y = 
#points for ONLY the US wines. Save your work to a jupyter ipynb.

df = pd.DataFrame(data)
usonly = df[df['country'] == 'US']
y = usonly.price.values
x = usonly.points.values

xaxis = list(x)
yaxis = list(y)

plt.title("plot chart")
plt.xlabel("price")
plt.ylabel("points")
plt.scatter(xaxis, yaxis)
plt.show()


 
```
